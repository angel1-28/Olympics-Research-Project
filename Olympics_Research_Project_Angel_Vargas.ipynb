```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("/athlete_events.csv")
df = data.dropna(subset=["Age", "Medal"])
data.head(10)
```

```python
# Here I only include the countries I will be focusing on for the project
target_countries = ['United States', 'Germany', 'Soviet Union', 'Japan', 'Jamaica', 'Russia', 'East Germnay', 'West Germany']

# Filtering it
filtered_df = df[df['Team'].isin(target_countries)]
filtered_df.head(20)
```

```python
# Here I grouped the medalists by year and got the age average
avg_age_by_year = df.groupby('Year')['Age'].mean().reset_index()
avg_age_by_year.head(20)
```

```python
plt.figure(figsize=(12, 6))
plt.plot(avg_age_by_year['Year'], avg_age_by_year['Age'], marker='o', linestyle='-')
plt.title('Average Age of Medalists Over Time')
plt.xlabel('Year')
plt.ylabel('Average Age')
plt.grid(True)
plt.show()
```

```python
# this shows the top 10 countries with the most wins
medals_df = data.loc[data['Medal'].isin(['Gold', 'Silver', 'Bronze'])]
medals_df.head(10)

total_medals = medals_df['NOC'].value_counts().reset_index()
total_medals.columns = ['Country', 'Total Medals']
total_medals.head(10)
```

```python
plt.bar(total_medals['Country'].head(5), total_medals['Total Medals'].head(5))
plt.xlabel('Country')
plt.ylabel('Total Medals')
plt.title('Total Medals by Country (Top 5)')
total_medals.head(10)
plt.show()
```

```python
# shows how many medals each country won per year
medals_per_year = medals_df.groupby(['Year', 'NOC']).size().reset_index(name='Medals')
medals_per_year.head(10)
```
